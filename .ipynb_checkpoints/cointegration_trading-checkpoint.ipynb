```python
import numpy as np
import pandas as pd
import yfinance as yf
import statsmodels.api as sm
import scipy.optimize as spop
import matplotlib.pyplot as plt
%matplotlib inline
import warnings

warnings.simplefilter('ignore')
```

```python
# Definindo os parâmetros da estratégia
stocks = ['ASAI3.SA', 'RENT3.SA']
start = '2023-01-02'
end = '2023-08-23'
window = 20
t_threshold = -2.5

data = pd.DataFrame()
returns = pd.DataFrame()

for stock in stocks:
    prices = yf.download(stock, start, end)
    data[stock] = prices['Close']
    returns[stock] = np.append(data[stock][1:].reset_index(drop=True)/data[stock][:-1].reset_index(drop=True) - 1, 0)
```

```python
for stock in data.columns:
    data[f'return_{stock}'] = data[stock].pct_change()
```

```python
data.head()
```

```python
data.dropna(inplace=True)
```

```python
gross_returns = np.array([])
t_s = np.array([])
s_1 = stocks[0]
s_2 = stocks[1]
```

```python
data.head()
```

```python
# Inicialização das variáveis
gross_returns = np.array([])  # Array para armazenar os retornos brutos
t_s = np.array([])  # Array para armazenar os valores de t dicker-fuller
s_1 = stocks[0]  # Primeira ação
s_2 = stocks[1]  # Segunda ação

# Loop através dos dias
for t in range(window, len(data)):
    # Definição da função para calcular a raiz unitária
    def unit_root(b):
        a = np.average(data[s_2][t-window:t] - b * data[s_1][t-window:t])  # Cálculo de 'a' na equação de cointegração
        fair_value = a + b * data[s_1][t-window:t]  # Cálculo do preço justo
        diff = np.array(fair_value - data[s_2][t-window:t])
        diff_diff = diff[1:] - diff[:-1]
        linreg = sm.OLS(diff_diff, diff[:-1])
        res = linreg.fit()
        return res.params[0] / res.bse[0]

    # Otimização dos parâmetros da equação de cointegração
    res1 = spop.minimize(unit_root, data[s_1][t] / data[s_2][t], method='Nelder-Mead')
    t_opt = res1.fun
    b_opt = float(res1.x)
    a_opt = np.average(data[s_2][t-window:t] - b_opt * data[s_1][t-window:t])  # Cálculo de 'a' otimizado
    fair_value = a_opt + b_opt * data[s_1][t]  # Cálculo do preço justo otimizado
    
    # Verificação das condições para operar
    if t == window:
        old_signal = 0
    if t_opt > t_threshold:
        signal = 0
        gross_return = 0
    else:
        signal = np.sign(fair_value - data[s_2][t])
        gross_return = signal * (returns[s_2][t] - returns[s_1][t])

    # Armazenamento dos resultados
    gross_returns = np.append(gross_returns, gross_return)
    t_s = np.append(t_s, t_opt)
        
    # Impressão dos resultados
    print(f'Dia {data.index[t]}')
    print('--------------------------------')
    
    if signal == 0:
        print('Não operar')
    elif signal == 1:
        print(f'Entrar long em {s_2} e short em {s_1}')
    else:
        print(f'Entrar long em {s_1} e short em {s_2}')

    print(f'Retorno bruto: {round(gross_return * 100, 2)}%')
    print(f'Retorno bruto acumulado: {round(np.prod(1 + gross_returns) * 100 - 100, 2)} %')

# Plot dos resultados acumulados
plt.plot(np.append(1, np.cumprod(1 + gross_returns)))
```
